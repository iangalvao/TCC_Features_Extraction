<h1>Extração de Luminosidade Média</h1>

Esse notebook tem como objetivo a extração da característica luminosidade média de uma imagem.
Para realizar a extração é usada uma transformação no espaço de cor de RGB para Lab, o que permite acessar diretamente os valores de luminância da imagem. 


## Bibliotecas e Funções

```python
from matplotlib.pyplot import imshow
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from skimage.color import rgb2lab, lab2rgb
```

## Código da função de extração
A extração consiste em transformar a imagem para o espaço Lab e cálcular a média do primeiro canal, o de luminância.

```python
def averageLum(imArray):
    imageLab = rgb2lab(imArray)
    LChannel = imageLab[:, :, 0]
    averageLum = np.average(LChannel)
    return (averageLum)
```

## Exemplos
A seguir temos exemplos de imagens que demonstram os passos percorridos pela função de extração e ilustram o espaço de cor Lab. As imagens utilizadas se encontram na pasta <b>sample_data<b>.

```python
image1 = Image.open('../../sample_data/1024x768/test_images/5076506.jpg')
image2 = Image.open('../../sample_data/1024x768/test_images/145215893.jpg')
imArray1 = np.array(image1)
imArray2 = np.array(image2)

plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(imArray1), plt.axis('off'), plt.title('Imagem 1', size=10)
plt.subplot(122), plt.imshow(imArray2), plt.axis('off'), plt.title('Imagem 2', size=10)
plt.show()
```

### Transformação RGB para Lab e separação dos canais:

```python
imageLab1 = rgb2lab(imArray1)
lChannel1 = imageLab1[:,:,0]
aChannel1 = imageLab1[:,:,1]
bChannel1 = imageLab1[:,:,2]

imageLab2 = rgb2lab(imArray2)
lChannel2 = imageLab2[:,:,0]
aChannel2 = imageLab2[:,:,1]
bChannel2 = imageLab2[:,:,2]
```

## Canais de Cor no Espaço Lab
Abaixo são mostrados os 3 canais do espaço de cor Lab da imagem usada como exemplo. Note que os valores de máximo e mínimo para cada canal são diferentes dos do espaço RGB. Temos:
   - Luminância
       - min: 0 / max: 100
   - a
       - min: -127 / max: 127
   - b
       - min: -127 / max: 127

### Plot dos canais de luminância.

```python
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(lChannel1, cmap='gray', vmin=0, vmax=100), plt.axis('off'), plt.title('Imagem 1 - Luminâmcia', size=10)
plt.subplot(122), plt.imshow(lChannel2, cmap='gray', vmin=0, vmax=100), plt.title('Imagem 2 - Luminância', size=10)
plt.show()
```

### Plot dos canais a.

```python
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(aChannel1, cmap='gray', vmin=-127, vmax=127), plt.axis('off'), plt.title('Imagem 1 - Canal a', size=10)
plt.subplot(122), plt.imshow(aChannel2, cmap='gray', vmin=-127, vmax=127), plt.title('Imagem 2 - Canal a', size=10)
plt.show()
```

### Plot dos canais b.

```python
plt.figure(figsize=(10,5))
plt.subplot(121), plt.imshow(bChannel1, cmap='gray', vmin=-127, vmax=127), plt.axis('off'), plt.title('Imagem 1 - Canal b', size=10)
plt.subplot(122), plt.imshow(bChannel2, cmap='gray', vmin=-127, vmax=127), plt.title('Imagem 2 - Canal b', size=10)
plt.show()
```

## Features
Por último, temos as features extraídas das imagens:

```python
avLum1 = averageLum(imArray1)
avLum2 = averageLum(imArray2)
print (f"Luminância média da imagem 1: {avLum1}" )
print (f"Luminância média da imagem 2: {avLum2}" )
```
